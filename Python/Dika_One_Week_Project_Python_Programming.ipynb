```python
kelas = []

while True:
    data_kelas = {}

    print("=== Masukkan Data Kelas ===")
    nama_kelas = input("Nama Kelas: ")
    durasi_kelas = input("Durasi Kelas: ")
    nama_mentor = input("Nama Mentor: ")
    pertemuan = input("Pertemuan (online/offline): ")
    jumlah_pertemuan = int(input("Berapa kali pertemuan: "))
    kuota = int(input("Kuota Kelas: "))
    link_forum = input("Link Forum Kelas: ")
    aktif = input("Kelas Aktif (ya/tidak): ")

    data_kelas["Nama Kelas"] = nama_kelas
    data_kelas["Durasi Kelas"] = durasi_kelas
    data_kelas["Nama Mentor"] = nama_mentor
    data_kelas["Pertemuan"] = pertemuan
    data_kelas["Jumlah Pertemuan"] = jumlah_pertemuan
    data_kelas["Kuota"] = kuota
    data_kelas["Link Forum"] = link_forum
    data_kelas["Aktif"] = aktif.lower() == "ya"

    data_kelas["Murid"] = []

    opsi_murid = input("Apakah Anda ingin memasukkan data murid? (ya/tidak): ")
    if opsi_murid.lower() == "ya":
        selesai_murid = False
        while not selesai_murid:
            if len(data_kelas["Murid"]) < kuota:
                data_murid = {}
                print("=== Masukkan Data Murid ===")
                nama_murid = input("Nama Murid: ")
                email_murid = input("Email Murid: ")
                pekerjaan_murid = input("Pekerjaan Murid: ")
                nomor_whatsapp_murid = input("Nomor Whatsapp Murid: ")

                data_murid["Nama"] = nama_murid
                data_murid["Email"] = email_murid
                data_murid["Pekerjaan"] = pekerjaan_murid
                data_murid["Nomor Whatsapp"] = nomor_whatsapp_murid

                data_kelas["Murid"].append(data_murid)

                opsi_tambah_murid = input("Tambahkan data murid lagi? (ya/tidak): ")
                if opsi_tambah_murid.lower() == "tidak":
                    selesai_murid = True
            else:
                print("Kelas sudah mencapai kuota maksimal!")
                selesai_murid = True

    kelas.append(data_kelas)

    opsi = input("Tambahkan data kelas lagi? (ya/tidak): ")
    if opsi.lower() == "tidak":
        break

while True:
    print("=== Daftar Kelas ===")
    for data_kelas in kelas:
        print("- ", data_kelas["Nama Kelas"])
        print()

    print("Silahkan pilih opsi dibawah ini")
    print("1. Tambahkan Kelas")
    print("2. Filter Kelas")
    print("3. Keluar")

    opsi = input("Pilih opsi: ")
    if opsi == "1":
        continue
    elif opsi == "2":
        nama_kelas_filter = input("Masukkan nama kelas yang ingin difilter: ")
        print(f"=== Daftar Kelas dengan Nama {nama_kelas_filter} ===")
        kelas_ditemukan = False
        for data_kelas in kelas:
            if data_kelas["Nama Kelas"].lower() == nama_kelas_filter.lower():
                print("Nama Kelas:", data_kelas["Nama Kelas"])
                print("Durasi Kelas:", data_kelas["Durasi Kelas"])
                print("Nama Mentor:", data_kelas["Nama Mentor"])
                print("Pertemuan:", data_kelas["Pertemuan"])
                print("Jumlah Pertemuan:", data_kelas["Jumlah Pertemuan"])
                print("Kuota:", data_kelas["Kuota"])
                print("Link Forum:", data_kelas["Link Forum"])
                print("Kelas Aktif:", "Ya" if data_kelas["Aktif"] else "Tidak")

                if len(data_kelas["Murid"]) > 0:
                    print("Data Murid:")
                    for murid in data_kelas["Murid"]:
                        print("Nama:", murid["Nama"])
                        print("Email:", murid["Email"])
                        print("Pekerjaan:", murid["Pekerjaan"])
                        print("Nomor Whatsapp:", murid["Nomor Whatsapp"])

                kelas_ditemukan = True

        if not kelas_ditemukan:
            print("Tidak ada kelas dengan nama yang sesuai filter.")
    elif opsi == "3":
        break
    else:
        print("Opsi yang Anda pilih tidak valid. Silakan pilih opsi yang tersedia.")

    print()

print("Terima kasih telah menggunakan aplikasi ini!")
```
